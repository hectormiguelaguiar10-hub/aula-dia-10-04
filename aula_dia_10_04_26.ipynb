```python
#escrever um progrma que verifica as notas do 4 bimestre ,calcula a media do aluno

#nao permitir que o usuario digite notas menores que 0 ou 10
#media menor que 5- reprovado
#media maior que 5 e menor que 7 - recuperacao
# media maior que 7 aprovado



# entrada

# A linha abaixo foi removida ou comentada pois a variável 'nota' era ambígua e não era a média das bimestrais.
# nota = float(input("digite sua nota "))

# Função para obter uma nota válida entre 0 e 10
def get_valid_grade(bimestre_num):
    while True:
        try:
            grade = float(input(f"digite sua nota do {bimestre_num} bimestre: "))
            if 0 <= grade <= 10:
                return grade
            else:
                print ("Nota inválida. Por favor, digite sua nota de 0 a 10.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

b1 = get_valid_grade(1)
b2 = get_valid_grade(2)
b3 = get_valid_grade(3)
b4 = get_valid_grade(4)

#processamneto e saida

media = (b1 + b2 + b3 + b4) / 4

if media < 5:
  print(f"O aluno foi reprovado com média {media:.2f}")

elif media >= 5 and media < 7:
  print (f"O aluno ficou de recuperacao com média {media:.2f}")

else:
  print (f"O aluno foi aprovado com média {media:.2f}")
```
